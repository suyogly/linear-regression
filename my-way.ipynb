```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.loadtxt("/kaggle/input/weight-hight-dataset/weight-height (1).csv", delimiter=",", skiprows=1, usecols=(1, 2))
data
```

```python
data.size
```

```python
x = data[:, 0]/12
y = data[:, 1]/2.205
x[0], y[0]
```

```python
plt.plot(x, y, "o")
plt.xlabel("height")
plt.ylabel("weight")
plt.title("Ratio between height and weight")
plt.show()
```

```python
mean_x = np.mean(x)
print(f"average height: {mean_x} \n")

mean_y = np.mean(y)
print(f"average weight: {mean_y} \n")
```

```python
weight_per_height = mean_y/mean_x
weight_per_height
```
